### Begin by working with county data
#### We will get to tracts later!

```python
counties.info()
```

```python
counties.sort_values("PCT_LACCESS_POP15", ascending=False)
```

<p>So, I need to do some quick math to figure out population, instead of loading another dumb excel file in.</p>

```python
counties["population"] = counties["LACCESS_POP15"] /(counties["PCT_LACCESS_POP15"]/100)
```

```python
counties_sorted_pop = counties.sort_values("population", ascending=False)
counties_sorted_pop.head(5)
```

<p>Looks like it worked.<p>

```python
counties_sorted_food_access = counties.sort_values("PCT_LACCESS_LOWI15", ascending=False)
counties_sorted_food_access.head(5)
```

<p>Now we'll look at food insecurity in the top decile of U.S. counties.</p>
<p>We can tweak this later if we want to look at the top quartile, and reuse this process to look at large census tracts (though, if you understand census tracts, won't really mean much.)</p>

```python
counties_sorted_food_access_top_10_poppercent = counties_sorted_food_access["population"].quantile(q=0.9)
```

```python
big_counties = counties_sorted_food_access[counties_sorted_food_access["population"] >= counties_sorted_food_access_top_10_poppercent]
```

```python
big_counties.head(5)
```

<p>Now let's just look at the columns we need then output the data for Lisa!</p>

```python
final_counties = big_counties[["State", "County", "PCT_LACCESS_LOWI15"]]
```

```python
final_counties.rename(columns={"PCT_LACCESS_LOWI15": "percent_without_grocery_access"}).head()
```

```python
final_counties = final_counties.rename(columns={"PCT_LACCESS_POP15": "percent_without_grocery_access"})
```

```python
len(final_counties)
```

```python
final_counties.to_csv("output/final_counties.csv")
```

### Now let's do some similar analyses for Census tracts

```python
tract_variables.head(50)
```

```python
urban_tracts = tracts[tracts["Urban"] == 1]
rural_tracts = tracts[tracts["Urban"] == 0]
```

```python
urban_tracts = urban_tracts[["CensusTract", "State", "County", "lalowi1", "lalowi1share"]]
```

```python
urban_tracts = urban_tracts.sort_values("lalowi1", ascending=False)
```

```python
rural_tracts = rural_tracts[["CensusTract", "State", "County", "lalowi10", "lalowi10share"]]
```

```python
rural_tracts = rural_tracts.sort_values("lalowi10", ascending=False)
```

```python
florida_urban_tracts = urban_tracts[urban_tracts["State"] == "Florida"]
```

```python
florida_urban_tracts.head()
```

```python
florida_rural_tracts = rural_tracts[rural_tracts["State"] == "Florida"]
```

```python
florida_rural_tracts.head()
```

```python
rural_tracts = rural_tracts.rename(columns={"lalowi10": "pop_with_low_access", "lalowi10share": "percent_with_low_access"})
urban_tracts = urban_tracts.rename(columns={"lalowi1": "pop_with_low_access", "lalowi1share": "percent_with_low_access"})
florida_rural_tracts = florida_rural_tracts.rename(columns={"lalowi10": "pop_with_low_access", "lalowi10share": "percent_with_low_access"})
florida_urban_tracts = florida_urban_tracts.rename(columns={"lalowi1": "pop_with_low_access", "lalowi1share": "percent_with_low_access"})
```

```python
rural_tracts.to_csv("output/final_rural_tracts.csv")
urban_tracts.to_csv("output/final_urban_tracts.csv")
florida_rural_tracts.to_csv("output/final_florida_rural_tracts.csv")
florida_urban_tracts.to_csv("output/final_florida_urban_tracts.csv")
```
