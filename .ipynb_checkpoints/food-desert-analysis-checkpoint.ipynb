### Begin by working with county data
#### We will get to tracts later!

```python
counties.info()
```

```python
counties.sort_values("PCT_LACCESS_POP15", ascending=False)
```

<p>So, I need to do some quick math to figure out population, instead of loading another dumb excel file in.</p>

```python
counties["population"] = counties["LACCESS_POP15"] /(counties["PCT_LACCESS_POP15"]/100) = counties["LACCESS_POP15"] /(counties["PCT_LACCESS_POP15"]/100)
```

```python
counties_sorted_pop = counties.sort_values("population", ascending=False)
counties_sorted_pop.head(5)
```

<p>Looks like it worked.<p>

```python
counties_sorted_food_access = counties.sort_values("PCT_LACCESS_POP15", ascending=False)
counties_sorted_food_access.head(5)
```

<p>Now we'll look at food insecurity in the top quartile of U.S. counties.</p>
<p>We can tweak this later if we want to look at the top decile, and reuse this process to look at large census tracts (though, if you understand census tracts, won't really mean much.)</p>

```python
counties_sorted_food_access_top_25_poppercent = counties_sorted_food_access["population"].quantile(q=0.75)
```

```python
big_counties = counties_sorted_food_access[counties_sorted_food_access["population"] >= counties_sorted_food_access_top_25_poppercent]
```

```python
big_counties.head(50)
```
